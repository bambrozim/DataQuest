```python
def read_csv(file_str):
    string_list = open(file_str, "r").read().split("\n")[1:]
    final_list = []
    for row in string_list:
        int_fields = []
        string_fields = row.split(",")
        for value in string_fields:
            int_fields.append(int(value))
        final_list.append(int_fields)
    return final_list

cdc_list = read_csv("US_births_1994-2003_CDC_NCHS.csv")
cdc_list
```

```python
def month_births(input_lst):
    births_per_month = {}
    for row in input_lst:
        month = row[1]
        births = row[4]
        if month in births_per_month:
            births_per_month[month] += births
        else:
            births_per_month[month] = births
    return births_per_month

cdc_month_births = month_births(cdc_list)
cdc_month_births
```

```python
def dow_births(input_lst):
    births_per_dow = {}
    for row in input_lst:
        dow = row[3]
        births = row[4]
        if dow in births_per_dow:
            births_per_dow[dow] += births
        else:
            births_per_dow[dow] = births
    return births_per_dow

cdc_day_births = dow_births(cdc_list)
cdc_day_births
```

```python
def calc_counts(data, column):
    births_counts = {}
    for row in data:
        data_column = row[column]
        births = row[4]
        if data_column in births_counts:
            births_counts[data_column] += births
        else:
            births_counts[data_column] = births
    return births_counts

cdc_year_births = calc_counts(cdc_list, 0)
cdc_month_births = calc_counts(cdc_list, 1)
cdc_dom_births = calc_counts(cdc_list, 2)
cdc_dow_births = calc_counts(cdc_list, 3)
```

## Calculating the min and max for any dictionary

```python
def min_max(input_dict):
    max_value = max(input_dict.keys(), key = lambda k: input_dict[k])
    min_value = min(input_dict.keys(), key = lambda k: input_dict[k])
    min_max = [input_dict[min_value], input_dict[max_value]]
    return min_max

cdc_dow_births_min_max = min_max(cdc_day_births)
cdc_dow_births_min_max
```

## A function that extracts the same values across years and calculates the differences between consecutive values to show if number of births is increasing or decreasing

```python
def across_years(input_dict, column, column_value):
    years_column = {}
    for row in input_dict:
        data_column = row[column]
        year = row[0]
        births = row[4]
        if data_column == column_value:
            if year in years_column:
                years_column[year] += births
            else:
                years_column[year] = births
    return years_column

saturday_births = across_years(cdc_list, 3, 6)
saturday_births
```
