## Guided Project: Exploring Gun Deaths in the US

```python
import csv

with open("guns.csv", "r") as f:
    data = list(csv.reader(f))

data[:5]
```

```python
headers = data[0]
data = data[1:]
```

```python
years = [row[1] for row in data]
years
```

```python
year_counts = {}
for year in years:
    if year in year_counts:
        year_counts[year] += 1
    else:
        year_counts[year] = 1
        
year_counts
```

```python
import datetime

dates = [datetime.datetime(year = int(row[1]), month = int(row[2]), day = 1) for row in data]
dates[:5]
```

```python
date_counts = {}
for date in dates:
    if date in date_counts:
        date_counts[date] += 1
    else:
        date_counts[date] = 1
        
date_counts
```

```python
sex_counts = {}
for row in data:
    sex = row[5]
    if sex in sex_counts:
        sex_counts[sex] += 1
    else:
        sex_counts[sex] = 1

sex_counts
```

```python
race_counts = {}
for row in data:
    race = row[7]
    if race in race_counts:
        race_counts[race] += 1
    else:
        race_counts[race] = 1

race_counts
```

```python
intents = [row[3] for row in data]
races = [row[7] for row in data]

homicide_race_counts = {}
for i, race in enumerate(races):
    intent = intents[i]
    if intent == "Homicide":
        if race in homicide_race_counts:
            homicide_race_counts[race] += 1
        else:
            homicide_race_counts[race] = 1

race_per_hundredk = {}
for k,v in homicide_race_counts.items():
    race_per_hundredk[k] = (v / mapping[k]) * 100000

race_per_hundredk
```

#### Figure out the link, if any, between month and homicide rate.

```python
months = [row[2] for row in data]

homicide_month_counts = {}
for i, month in enumerate(months):
    intent = intents[i]
    if intent == "Homicide":
        if month in homicide_month_counts:
            homicide_month_counts[month] += 1
        else:
            homicide_month_counts[month] = 1

sorted(homicide_month_counts.items())
```

It appears that has less homicides in february, besides that there's no link between the homicide rates with months.

#### Explore the homicide rate by gender.

```python
genders = [row[5] for row in data]

homicide_gender_counts = {}
for i, gender in enumerate(genders):
    intent = intents[i]
    if intent == "Homicide":
        if gender in homicide_gender_counts:
            homicide_gender_counts[gender] += 1
        else:
            homicide_gender_counts[gender] = 1

homicide_gender_counts
```

#### Find out if gun death rates correlate to location and education.
